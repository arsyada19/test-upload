```python
import numpy as np
import matplotlib.pyplot as plt

#membuat rumus cosinus
y = np.cos([0, 30,90, 120, 150, 180, 210, 270])
x = [0, 30,90, 120, 150, 180, 210, 270]

#tampilkan rumus cosinus dengan matplotlib
plt.title("Plot Cosinus")
plt.plot(x, y)
plt.xlabel("Besar Sudut ")
plt.ylabel("Nilai Cosinus Sudut")
plt.grid()
plt.show()
def plot1():
    #membuat rumus cosinus
    y = np.cos([0, 30,90, 120, 150, 180, 210, 270])
    x = [0, 30,90, 120, 150, 180, 210, 270]

    #tampilkan rumus cosinus dengan matplotlib
    plt.title("Plot Cosinus")
    plt.plot(x, y)
    plt.xlabel("Besar Sudut ")
    plt.ylabel("Nilai Cosinus Sudut")
    plt.grid()
    plt.show()
```

```python
import pandas as pd
data = pd.read_csv("GooglePlayStore_wild.csv", encoding = "utf-8")


###### cdoing pertama
plt.title("Rating Vs Reviews")
plt.plot(data['Rating'], data['Reviews']) #plt.plot(sumbu x, sumbu y)
plt.xlabel("Rating")
plt.ylabel("Reviews")
#mengatifkan grid
plt.grid() #boleh dipakai boleh tidak
#### coding kedua
plt.title("Plot Rating")
plt.plot(data['Rating']) #tanpa sumbu x, otomatis menjadi sumbu y
plt.xlabel("nomor")
plt.ylabel("Rating")
```

```python
#grafik histogram
plt.hist(['Rating'])
plt.xlabel("Rating")
plt.title("Grafik persebaran Rating")
```

```python

#plot pie
data['Category'].value_counts().plot(kind = 'pie')
#####code pie plot versi 2
#menampilkan diagram pie dalam bentuk persen
plt.title("Pie Diagram")
sum_category = data['Category'].value_counts().sum() #hitung dlu jumlah semua data
(data['Category'].value_counts()/sum_category).plot(kind = 'pie', autopct="%1.1f%%") #jumlah tiap data/semua data x 100%
plt.show()
#plot bar chart #harus data berbetuk string
plt.title("Jumlah Masing-Masing Kategori")
data['Category'].value_counts().plot(kind = 'bar')
```
